```python
import pandas as pd


df = pd.read_excel("/kaggle/input/datasets/kirtisharma8433/project-1/Dataset for Data Analytics.xlsx")
```

```python
df.head()
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
df.duplicated().sum()
```

```python
df["OrderID"].duplicated().sum()
df["TrackingNumber"].duplicated().sum()
```

```python
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")
```

```python
print(df["Date"].head())
```

```python
text_cols = [
    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "CouponCode",
    "ReferralSource"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
num_cols = ["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["CalculatedTotal"] = df["Quantity"] * df["UnitPrice"]

mismatch = df[df["CalculatedTotal"].round(2) != df["TotalPrice"].round(2)]

print("\nTotalPrice Mismatch Count:", len(mismatch))
```

```python
df.drop(columns=["CalculatedTotal"], inplace=True)
```

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

print("\nFinal Duplicate Check:")
print("Duplicate Rows:", df.duplicated().sum())
print("Duplicate OrderID:", df["OrderID"].duplicated().sum())
print("Duplicate TrackingNumber:", df["TrackingNumber"].duplicated().sum())

print("\nFinal Info:")
print(df.info())
```

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)

print("Dataset exported successfully!")
```
